## Annotations of the Issues Data Presented ##

### **Data Issues Identified Before Cleansing** ###

**1. Inconsistent Column Names**  
Some columns include spaces and capital letters (e.g., “Customer Type”), which can cause errors in referencing column names during analysis.

**2. Inconsistent Categorical Values**  
- *Gender:* Includes inconsistent entries such as “F”, “M”, and “Fe male”.  
- *Customer Type:* Variants like “Loyal Customer”, “Disloyal Customer”, and “disloyal” need to be standardized.  
- *Type of Travel:* Contains “Business travel”, “Personal Travel”, and “Business”.

**3. Outliers and Invalid Numeric Data**  
- *Age:* Includes unrealistic values like 7 and 185.  
- *Flight Distance:* Contains extreme outliers (e.g., 2,820,000 and 9,000,000).

**4. Improper Rating Scales**  
Rating columns (Inflight WiFi Service, Cleanliness, etc.) contain text entries like “Great” and invalid values like 0 or 10.  
The valid range should be **1–5**.

**5. Mixed Data Types**  
Some numeric columns are stored as objects due to the presence of text or symbols.

**6. Missing Values (NaN)**  
Several columns, including *Ease of Online Booking*, *Food and Drink*, *Inflight Entertainment*, *Baggage Handling*, *Cleanliness*, and *Age*, have missing values.

**7. Duplicated Data**  
The dataset contains 3,864 duplicated rows, which must be removed to prevent bias.

**8. Ambiguous Satisfaction Labels**  
The *Satisfaction* column contains inconsistent values such as “satisfied”, “Satisfied”, “neutral”, and numeric codes like 0 and 1, requiring standardization.


```python
###CONDUCT DATA CLEANSING###
##Verify df_Business Columns
for col in df_Business.columns:
    print(repr(col))
```

```python
##Verify columns from the original Database
for c in df.columns: print(repr(c))
```

```python
##Drop 'Class' Column
df_Business = df_Business.drop(columns=['Class'], errors='ignore')
```

```python
#Verify column was dropped
df_Business.columns
```

```python
##Rename Columns in Lowercase and Remove Space and Replace it with _
df_Business.rename(columns = lambda x: x.lower().replace(' ','_'), inplace= True)
df_Business.columns
```

```python
##Modify Improperly Formatted Data and Handle Bad Data
#Column 1 'Gender' - Standardized abbreviations ("F"/"M") and spacing issues ("Fe male") to "Female"/"Male"
df_Business['gender'] = df_Business['gender'].astype(str).str.strip()
df_Business['gender'] = df_Business['gender'].replace({
    'F':'Female','M':'Male','Fe male':'Female'})

```

```python
df_Business['gender'].unique()
```

```python
#Column 2 'customer_type' - Unified similar categories ("Loyal Customer"/"Disloyal Customer"/"disloyal") into "Loyal"/"Disloyal"
df_Business['customer_type'] = df_Business['customer_type'].replace({
    'Loyal Customer':'Loyal',
    'Disloyal Customer':'Disloyal',
    'disloyal':'Disloyal'})
```

```python
df_Business['customer_type'].unique()
```

```python
import numpy as np
```

```python
#Column 3 'age' - Ensuring age fits a reasonable range 
df_Business.loc[~df_Business['age'].between(12, 90), 'age'] = np.nan
```

```python
df_Business['age'].unique()
```

```python
#Column 4 'type_of_travel' - Standardized travel categories ("Business"-"Business Travel") for naming consistency 
df_Business['type_of_travel'] = df_Business['type_of_travel'].replace({
    'Business':'Business travel',
    'Personal Travel':'Personal Travel'})
```

```python
df_Business['type_of_travel'].unique()
```

```python
#Column 5 'flight_distance' - Get rid of outiers in flight distance 
df_Business.loc[df_Business['flight_distance'] <= 0, 'flight_distance'] = np.nan
lo, hi = df_Business['flight_distance'].quantile([0.01, 0.99])
df_Business['flight_distance'] = df_Business['flight_distance'].clip(lo, hi)
```

```python
df_Business['flight_distance'].unique()
```

```python
#Columns 6-13 - Rate ordinal values (0-5) 
rating_rest = [
    'inflight_wifi_service',
    'departure/arrival_time_convenient',
    'ease_of_online_booking',
    'food_and_drink',
    'inflight_entertainment',
    'baggage_handling',
    'cleanliness'
]

rating_rest = [c for c in rating_rest if c in df_Business.columns]

for c in rating_rest:
    df_Business[c] = pd.to_numeric(df_Business[c], errors='coerce')
    df_Business.loc[~df_Business[c].between(0, 5), c] = np.nan
```

```python
#Columns 6-13 - Adjust rating scale from 0-5 to 1-5; converting 0 to NaN
rating_cols = [
    'inflight_wifi_service',
    'departure/arrival_time_convenient',
    'ease_of_online_booking',
    'food_and_drink',
    'inflight_entertainment',
    'baggage_handling',
    'cleanliness'
]


for c in rating_cols:
    df_Business[c] = pd.to_numeric(df_Business[c], errors='coerce')
    df_Business.loc[(df_Business[c] == 0) | (~df_Business[c].between(1, 5)), c] = np.nan
```

```python
###Check the corect columns were cleaned   
rating_rest
```

```python
###Check each cleaned columns unique values; to make sure no out-of-range values
for c in rating_rest:
    print(f"\n{c} unique values:")
    print(sorted(df_Business[c].dropna().unique()))
```

```python
###Check how many missing values were created after cleaning
df_Business[rating_rest].isna().sum()
```

```python
###Confirm numeric Integrity - min (1) max (5); Show a descriptive summary of the cleaned columns
df_Business[rating_rest].describe()
```

```python
#Column 14 'departure_delay_in_minutes'- Convert into numbers, convert value less then zero into NaN, get rid of extreme value
for c in ['departure_delay_in_minutes', 'arrival_delay_in_minutes']:
    df_Business[c] = pd.to_numeric(df_Business[c], errors='coerce')
    df_Business.loc[df_Business[c] < 0, c] = np.nan
    lo, hi = df_Business[c].quantile([0.01, 0.99])
    df_Business[c] = df_Business[c].clip(lo, hi)

```

```python
###Verify result 
df_Business[['departure_delay_in_minutes', 'arrival_delay_in_minutes']].describe()
```

```python
#Column 15 'Satisfaction' - Reconstruct the label 
df_Business['satisfaction'] = (
        df_Business['satisfaction']
          .replace({'satisfied':'Satisfied',
                    'dissatisfied':'Dissatisfied',
                    'neutral':'Neutral',
                    '1':'Satisfied',
                    '0':'Dissatisfied'}))
```

```python
##Verify the Output
df_Business['satisfaction'].unique()
```

```python
##Handle Missing Values
#Categorical Values - Fill with 'Unknown'
for col in ['customer_type', 'type_of_travel', 'gender']:
    if col in df_Business.columns:
        df_Business[col] = df_Business[col].fillna('Unknown')
```

```python
#Continuos Numeric Values - Fill with median
for col in ['age', 'flight_distance']:
    if col in df_Business.columns:
        df_Business[col] = df_Business[col].fillna(df_Business[col].median())
```

```python
#Rating Columns (0-5) - Fill with mode
rating_cols = [
    'inflight_wifi_service',
    'departure/arrival_time_convenient',
    'ease_of_online_booking',
    'food_and_drink',
    'inflight_entertainment',
    'baggage_handling',
    'cleanliness'
]
rating_cols = [c for c in rating_cols if c in df_Business.columns]

for col in rating_cols:
    mode_val = df_Business[col].mode(dropna=True).iat[0]
    df_Business[col] = df_Business[col].fillna(mode_val)
```

```python
#Delay Columns - Fill with median 
for col in ['departure_delay_in_minutes', 'arrival_delay_in_minutes']:
    if col in df_Business.columns:
        df_Business[col] = df_Business[col].fillna(df_Business[col].median())
```

```python
#Satisfaction - Fill with "Unknown"
if 'satisfaction' in df_Business.columns:
    df_Business['satisfaction'] = df_Business['satisfaction'].fillna('Unknown')
```

```python
##Verify that no missing values remain 
print(df_Business.isna().sum().sort_values(ascending=False).head(10))
print("Total NA:", int(df_Business.isna().sum().sum()))
```

```python
##Remove fully duplicated observations
df_Business = df_Business.drop_duplicates()
```

```python
#Verify
print("Duplicates after drop:", df_Business.duplicated().sum())
print("Final row count:", len(df_Business))
```

```python
##Fix Column Datatypes
#Convert rating columns from object to float
cols_to_float = ['ease_of_online_booking', 'food_and_drink', 'baggage_handling']
for col in cols_to_float:
    if col in df_Business.columns:
        df_Business[col] = df_Business[col].astype(float)
```

```python
#Convert categorical values to category dtype
cat_cols = ['gender', 'customer_type', 'type_of_travel', 'satisfaction']
for col in cat_cols:
    if col in df_Business.columns:
        df_Business[col] = df_Business[col].astype('category')
```

## Final Annotations — Comparing the Data Before and After Cleansing 

The original dataset went through multiple data cleansing processes to ensure **consistency, accuracy, and readiness for analysis**.

---

### **Summary of Column Changes**

**Gender**  
Originally stored as text (`object`) with inconsistent entries such as “Fe male”, “M”, and “F”. These were standardized to “Male” and “Female”, converted to categorical type, and all missing values were filled with **'Unknown'**.

**Customer_Type**  
Contained mixed formats (“Loyal Customer”, “Disloyal”, and lowercase variants). Categories were unified into **“Loyal”** and **“Disloyal”**, datatype converted to categorical, and missing values filled with **'Unknown'**.

**Age**  
Contained out-of-range values (7 and 185) and some missing entries. Values outside the realistic range (12–90) were replaced with NaN, then filled with the **median age**. Datatype remained numeric (`float`).

**Type_of_Travel**  
Included inconsistent categories (“Business”, “Business travel”) and missing data. These were standardized to **“Business travel”** and **“Personal Travel”**, converted to categorical, and missing values replaced with **'Unknown'**.

**Flight_Distance**  
Contained extreme outliers and missing values. Outliers were capped within the **1st and 99th percentiles**, datatype preserved as numeric (`float`), and missing entries filled with the **median**.

**Inflight_Wifi_Service**  
Originally numeric but included values outside the 1–5 range. Values were restricted to valid ratings (**1–5**), converted to numeric, and missing values filled with the **mode**.

**Departure/Arrival_Time_Convenient**  
Contained out-of-range and missing values. All values were limited to the **1–5 range**, datatype set to `float`, and missing entries replaced with the **mode**.

**Ease_of_Online_Booking**  
Read as object type due to text and “Unknown” entries. Converted to `float`, cleaned of non-numeric values, restricted to **1–5 range**, and filled with the **mode**.

**Food_and_Drink**  
Contained text entries like “Great” and numeric strings. Converted to numeric, restricted to the **1–5 range**, and missing values filled with the **mode**.

**Inflight_Entertainment**  
Included invalid ratings (e.g., 10) and missing values. Out-of-range values were replaced with NaN, restricted to **1–5**, datatype set to `float`, and missing entries filled with the **mode**.

**Baggage_Handling**  
Contained invalid entry “A” and some missing values. Converted to `float`, invalid entries replaced with NaN, restricted to **1–5**, and missing data filled with the **mode**.

**Cleanliness**  
Numeric but with some missing entries and minor inconsistencies. Restricted to the **1–5 range**, missing values filled with **mode**, and datatype kept as `float`.

**Departure_Delay_in_Minutes**  
Included negative and extreme delay times. Converted to numeric, negative values replaced with NaN, outliers capped within the **1st–99th percentile range**, and missing entries filled with the **median**.

**Arrival_Delay_in_Minutes**  
Similar to departure delay, negative and extreme values were cleaned, datatype converted to `float`, and missing entries replaced with the **median**.

**Satisfaction**  
Contained inconsistent labels (“1”, “0”, “satisfied”, “neutral”) and missing values. All categories standardized to **“Satisfied”**, **“Dissatisfied”**, and **“Neutral”**, datatype changed to categorical, and missing entries filled with the **most frequent label**.

**Class**  
This column was dropped as it was no longer needed for analysis.

---

### **Post-Cleansing Summary**

After cleansing, all columns are standardized to their correct datatypes, inconsistent entries were corrected, and **no missing values remain in the dataset**.  
The data is now **clean, consistent, and ready for analysis**.
**Overall Comparison Summary:**  
The data cleansing process successfully improved dataset quality by eliminating inconsistencies, correcting data types, and handling missing values. Compared to the original dataset, the cleaned version is fully standardized, reliable, and ready for analytical modeling.
